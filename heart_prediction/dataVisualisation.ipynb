### Basic statistic information

- Colinearity

```python
sns.heatmap(df.corr(), cmap="Blues", annot=True)
plt.title("Correlation heatmap between features")
plt.show()
```

- Data values dispertion

```python
def hist(values : pd.Series, title : str = "") -> None:
    """
    @values: an iterable variable that contains values to use to create the histogram
    """
    sns.set_style("whitegrid")
    sns.histplot(values)
    plt.title(title)
    plt.show()

def barplot(dataFrame : pd.DataFrame, x : str ,title : str = "") -> None:
    """
    @values: an iterable variable that contains values to use to create the histogram
    """
    sns.set_style("whitegrid")
    sns.countplot(data= dataFrame, x = x)
    plt.title(title)
    plt.show()
```

```python
hist(df["Age"], "Age histogram")
```

```python
hist(df["BloodPressure"], "Blood pressure histogram")
```

```python
hist(df["Cholesterol"], "Cholesterol histogram")
```

```python
hist(df["HeartRate"], "Heart rate histogram")
```
